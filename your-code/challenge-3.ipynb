## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: groupby to aggregate and then sum the grouped values
unique = orders.groupby(['CustomerID']).sum()
unique = unique.sort_values('amount_spent')
unique[['amount_spent']].head()
```

```python
# Sub Problem 2: identify which quartile represents which variable
Preferred_Customers = np.quantile(unique['amount_spent'],0.75)
VIP = np.quantile(unique['amount_spent'],0.95)

print(VIP)
print(Preferred_Customers)
```

```python
# Sub Problem 3: used the loc to find only the ones that meet the condition and update the 'Customer Label' column
unique.loc[unique['amount_spent'] > unique['amount_spent'].quantile(0.75), 'Customer Label'] = 'PC'
unique.loc[unique['amount_spent'] > unique['amount_spent'].quantile(0.95), 'Customer Label'] = 'VIP'
unique.head(10)
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
unique.loc[unique['Customer Label'] == 'VIP', 'Country'] = orders['Country']
VIP_count = unique[unique['Customer Label'] == 'VIP']['Country'].value_counts()
VIP_count = pd.DataFrame(VIP_count)
VIP_count
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
unique.loc[unique['Customer Label'] == 'PC', 'Country'] = orders['Country']
Preferred_count = unique[unique['Customer Label'] == 'PC']['Country'].value_counts()
Preferred_count = pd.DataFrame(Preferred_count)
Preferred_count
```

```python
VIP_count.add(Preferred_count, fill_value=0).astype(int)
```
